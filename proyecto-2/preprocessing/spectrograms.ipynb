# Generacion de espectrogramas a partir de los audios

```python
import librosa
import os
import matplotlib.pyplot as plt
import numpy as np

# Ruta donde está tu dataset
data_path = "../data/audio/"

# Cargar un archivo de audio (Ejemplo: archivo de audio de la clase '0')
file_path = os.path.join(data_path, '01', '0_01_0.wav')
audio, sr = librosa.load(file_path, sr=None)

print("Audio shape:", audio.shape)
print("Sample rate:", sr)

# Convertir audio a espectrograma (Mel)
spectrogram = librosa.feature.melspectrogram(y=audio, sr=sr, n_mels=128)

# Convertir a escala logarítmica para mejor visualización
log_spectrogram = librosa.power_to_db(spectrogram, ref=np.max)

# Mostrar el espectrograma
plt.figure(figsize=(10, 4))
plt.imshow(log_spectrogram, aspect='auto', origin='lower', cmap='inferno')
plt.title("Spectrogram")
plt.xlabel("Frames")
plt.ylabel("Mel Bands")
plt.colorbar(format='%+2.0f dB')
plt.show()
```
